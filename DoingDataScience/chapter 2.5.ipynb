# 불필요한 로우 제거

```python
df2 = data_frame.drop(range(4))
df2
```

# 컬럼 이름 정리하기
- 네번째 row에 있는 값들이 컬럼 값인 것 같으니 바꿔치자

### 네번째 row 값 출력

```python
list(data_frame.iloc[3].items())
```

```python
new_column_names = [v for k, v in data_frame.iloc[3].items()]
```

```python
df2.columns
```

### 네번째 row 값와 현재 컬럼을 매핑

```python
new_column_dict = dict(zip(df2.columns, new_column_names))
new_column_dict
```

### 업데이트

```python
df3 = df2.rename(columns=new_column_dict)
df3
```

```python
df3.sort_values(by=['SALE DATE'])
```
